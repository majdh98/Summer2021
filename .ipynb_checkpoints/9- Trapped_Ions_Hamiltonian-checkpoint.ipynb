# Trapped Ions Hamiltonian
In this notebook, we use qutip to simulate a Trapped Ions hamiltonian. We start by two qubits, but we provide a method that can simulate an arbitrary number of qubits. The Hamiltonian is of the form:
\begin{equation}H = -\sum_{i,j}J_{i,j}\sigma_i^z \sigma_j^z - h \sum\sigma_j^x  \end{equation}
or (for the case when J and h have 1 for every component)
\begin{equation} H = - \sigma^z \otimes \sigma^z - \sigma^z \otimes \sigma^z  -\sigma^x \otimes I - I \otimes \sigma^x \end{equation}


$J$ is an $N \times N$ matrix and $h$ is a one column-N row matrix where N is the number of qubits.

We compare the qutip simulation to the numerical solution where we use np.exp() to exponentiate the hamiltonian. 

```python
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
from scipy.linalg import expm

from scipy.integrate import ode

# for analytical exponentiation
from scipy.linalg import expm 
print("done")
```

# Qutip

```python
def construct_hamiltonian(J, h):
    N = len(J)
    
    sum_sx = []
    # build h times sum sigma_j^x
    for i in range(N):
        # make a list of N qeye matrices for one qubit
        op_list = []
        for _ in range(N):
            op_list.append(qeye(2))  
        # replace the jth element with sigmax, tensor, then append sum_sx
        op_list[i] = sigmax()
        sum_sx.append(h*tensor(op_list))
        
    sum_sz_sz = []
    # bild sum J_{i,j} sigma_i^z sigma_j^z, we ignore the case where i = j
    for i in range(N):      
        for j in range(N):
            if i == j:
                continue
            # make a list of N qeye matrices for one qubit
            op_list = []
            for _ in range(N):
                op_list.append(qeye(2))  
            # replace the ith and jth elements with sigma_z, tensor, then append to sum_sz_sz
            op_list[i] = sigmaz()
            op_list[j] = sigmaz()
            sum_sz_sz.append(J[i][j]*tensor(op_list))
    H = 0
    for op in sum_sz_sz:
        H -= op
    for op in sum_sx:
        H -=op
    return H
         
        
h = 1
J = np.array([[1, 1], [1, 1]])
construct_hamiltonian(J, h)
```

```python
# Simulation params

initial_state = tensor(basis(2, 0), basis(2, 0))

# define time points
times = np.linspace(0.0, 10, 100)

h = 1
J = np.array([[1, 1], [1, 1]])
H = construct_hamiltonian(J, h)
```

Run simulation:

```python
# run the simulation- to get expectation values we pass s_project
# define state on which to project for expectations
s_project = [tensor(sigmaz(), qeye(2)), tensor(qeye(2), sigmaz())]
result = sesolve(H, initial_state, times, s_project)# add progress_bar=True to see progress
expect_sz = result.expect

s_project = [tensor(sigmax(), qeye(2)), tensor(qeye(2), sigmax())]
result = sesolve(H, initial_state, times, s_project)# add progress_bar=True to see progress
expect_sx = result.expect

# run the simulation- to get states values we leave s_project as none
options = Options()
result = sesolve(H, initial_state, times)
qutip_states = result.states # used later to approximate error in estimation
print("done")
```

Unpack results:

```python
# expectation plot for sz
p = plt.figure(1)
p.set_figwidth(20)
p.set_figheight(8)
p = plt.subplot()
p.plot(times, expect_sz[0]) 
p.plot(times, expect_sz[1]) 
p.legend(("First Qubit", "Second Qubit",))
p.set_xlabel('Time') 
p.set_ylabel('Expectation values') 
p.set_title("Expectation values for sigma z for first qubit and second qubit")

# expectation plot for sx
p = plt.figure(2)
p.set_figwidth(20)
p.set_figheight(8)
p = plt.subplot()
p.plot(times, expect_sx[0]) 
p.plot(times, expect_sx[1]) 
p.legend(("First Qubit", "Second Qubit",))
p.set_xlabel('Time') 
p.set_ylabel('Expectation values') 
p.set_title("Expectation values for sigma x for first qubit and second qubit")
```

# Numerical Solution

Schrodinger equation states that:
$$ |\phi(t)\rangle = U|\phi (0)\rangle$$
where $ U = e^{-\frac{-iHt}{\hbar}}$

H in this case is not diagonal, becuase the $\sigma_x$ terms are not diagonal. We will use np.expm() to find the state of the system at the same time steps as in the qutip solution.

```python
# we can get H as an array from the qutip object we constructed before by using np.asarray()
H_numerical = np.asarray(H)

initial_state_numerical = np.asarray(initial_state)
# print(initial_state_numerical)
# print(initial_state)

numerical_states = [initial_state_numerical]
for t in times:
    if t == 0: 
        continue
    U = expm(-1j*H_numerical*t)
    mult = np.matmul(U, initial_state_numerical)
    numerical_states.append(mult)#/np.linalg.norm(mult))
```

Find error:

```python
error = []
for i in range(len(times)):
    y = []
    y.append(qutip_states[i][0] - numerical_states[i][0])
    y.append(qutip_states[i][1] - numerical_states[i][1])
    y.append(qutip_states[i][2] - numerical_states[i][2])
    y.append(qutip_states[i][3] - numerical_states[i][3])
    error.append(np.abs(np.vdot(y, y)))

p = plt.figure(1)
p.set_figwidth(20)
p.set_figheight(8)
p = plt.subplot()
p.plot(times, error) 
p.legend(("Qutip Error",))
p.set_xlabel('Time') 
p.set_ylabel('Error') 
p.set_title("Error in qutip for the trapped ions hamiltonian against numerical solution")
```
